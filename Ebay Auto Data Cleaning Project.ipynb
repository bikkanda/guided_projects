## Exploring Price by Brand

```python
autos["brand"].value_counts().head(20)
```

```python
top_20_brands = autos["brand"].value_counts().head(20).index
print(top_20_brands)
```

```python
brand_mean_prices = {}

for b in top_20_brands:
    selected_rows = autos[autos["brand"] == b]
    mean = selected_rows["price"].mean()
    brand_mean_prices[b] = int(mean)
    
brand_mean_prices    
```

```python
sorted_brands_mean_prices = sorted(brand_mean_prices.items(), reverse=True)
dict(sorted_brands_mean_prices)
```

```python
brand_mean_mileage = {}

for b in top_20_brands:
    selected_rows = autos[autos["brand"] == b]
    mean = selected_rows["odometer_km"].mean()
    brand_mean_mileage[b] = int(mean)
    
brand_mean_mileage
```

```python
bmp_series = pd.Series(sorted_brands_mean_prices)
print(bmp_series)
```

```python
bmm_series = pd.Series(brand_mean_mileage)
print(bmm_series)
```

```python
df = pd.DataFrame(bmp_series, columns=["mean_price"])
df
```

```python
new_df = df.assign(mean_mileage=bmm_series)
```
